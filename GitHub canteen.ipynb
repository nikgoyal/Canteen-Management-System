```python
from tkinter import *
import tkinter as tk
from tkinter import PhotoImage
from tkinter import Toplevel, messagebox, ttk, LEFT, Y, BOTTOM, X, W
import mysql.connector
from mysql.connector import Error
import customs as cs
from tkinter import ttk, messagebox
import re
import os
from datetime import datetime
from tkinter import simpledialog

# Initialize the main window
root = tk.Tk()
root.geometry("1030x706")
root.resizable(False,False)
root.title("C@nteeN")

# Load the image
try:
    bg = PhotoImage(file="food5.png")
except Exception as e:
    print(f"Error loading image: {e}")
    root.destroy()
    exit()

# Create a label with the image
label17 = Label(root, image=bg)
label17.place(x=0, y=0)

def update_datetime():
    current_datetime = datetime.now().strftime("%d-%m-%Y \n %H:%M:%S")
    datetime_label.config(text=current_datetime)
    root.after(1000, update_datetime)  # Update time every 1000 milliseconds (1 second)

# Create a frame to hold the datetime label and position it at the top-right
datetime_frame = tk.Frame(root, bg="#80471c")
datetime_frame.pack(side=tk.TOP, anchor='ne', padx=10, pady=10)

# Create a label to display current date and time
datetime_label = tk.Label(datetime_frame, text="", font="italic 15 bold", bg="#A0522D", fg="white", padx=10, pady=5, borderwidth=1, relief="solid")
datetime_label.pack()

# Start updating datetime initially
update_datetime()

label = Label(root, text="WeLcOmE to Our DOon C@nteeN", font="italic 20 bold", bg="#2E8B57", fg="white",padx=10, pady=5, borderwidth=1, relief="solid")
label.pack(pady=10)

# Shared list to store selected items
selected_items = []

def add_item(item):
    selected_items.append(item)
    messagebox.showinfo("Added", f"Added {item} to order list", parent=root)
    

def recommended_items():
    items = Toplevel(root)
    items.title("Recommended Menu")
    items.geometry("700x600")
    items.configure(bg="#80471c")

    # Load the background image
    bg_image = PhotoImage(file="food3.png")

    # Create a label to hold the background image
    bg_label = Label(items, image=bg_image)
    bg_label.place(relwidth=1, relheight=1)

    label = Label(items, text="Recommended Menu", font="italic 20 bold", bg="#80471c", fg="white",padx=10, pady=5, borderwidth=1, relief="solid")
    label.pack(pady=10)

    frame = Frame(items, bg="#80471c")
    frame.pack(pady=20)

    # Column headers
    header_id = Label(frame, text="R_No", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_name = Label(frame, text="R_Food", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_cost = Label(frame, text="Cost ₹", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    
    header_id.grid(row=0, column=0)
    header_name.grid(row=0, column=1)
    header_cost.grid(row=0, column=2)

    try:
        conn = mysql.connector.connect(host='localhost', user='Your Username', password='Your Password', database='Database name')
        cursor = conn.cursor()
        cursor.execute("select * from recommended;")
        rows = cursor.fetchall()
        conn.close()

        if not rows:
            messagebox.showinfo("Database Empty", "There is no data to show", parent=items)
        else:
            for idx, (r_id, r_name, r_cost) in enumerate(rows, start=1):
                label_id = Label(frame, text=r_id, bg="#80471c", fg="white", padx=10, pady=5)
                label_name = Label(frame, text=r_name, bg="#80471c", fg="white", padx=10, pady=5)
                label_cost = Label(frame, text=r_cost, bg="#80471c", fg="white", padx=10, pady=5)
                add_button = Button(frame, text="Add", command=lambda r_name=r_name, 
                            r_cost=r_cost: add_item(f"{r_name} ₹{r_cost}"), bg="#4CAF50", fg="white")
                
                label_id.grid(row=idx, column=0, sticky="w")
                label_name.grid(row=idx, column=1, sticky="w")
                label_cost.grid(row=idx, column=2, sticky="w")
                add_button.grid(row=idx, column=3, sticky="w")
                
    except Exception as e:
        messagebox.showerror("Error!", f"Error due to {str(e)}", parent=items)
        return

    # Keep a reference to avoid garbage collection
    items.bg_image = bg_image
    
def indian_veg():
    veg=Toplevel(root)
    veg.title("Indian Main Course Items")
    veg.geometry("700x600")
    veg.configure(bg="#F0E68C")
    
    # Load the background image
    bg_image = PhotoImage(file="food3.png")

    # Create a label to hold the background image
    bg_label = Label(veg, image=bg_image)
    bg_label.place(relwidth=1, relheight=1)

    label = Label(veg, text="Indian Main Course", font="italic 20 bold", bg="#80471c", fg="white",padx=10, pady=5, borderwidth=1, relief="solid")
    label.pack(pady=10)

    frame = Frame(veg, bg="#80471c")
    frame.pack(pady=20)

    # Column headers
    header_id = Label(frame, text="I_No", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_name = Label(frame, text="I_Food", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_cost = Label(frame, text="Cost ₹", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    
    header_id.grid(row=0, column=0)
    header_name.grid(row=0, column=1)
    header_cost.grid(row=0, column=2)

    try:
        conn = mysql.connector.connect(host='localhost', user='Your Username', password='Your Password', database='Database name')
        cursor = conn.cursor()
        cursor.execute("select * from indian_veg;")
        rows = cursor.fetchall()
        conn.close()

        if not rows:
            messagebox.showinfo("Database Empty", "There is no data to show", parent=items)
        else:
            for idx, (r_id, r_name, r_cost) in enumerate(rows, start=1):
                label_id = Label(frame, text=r_id, bg="#80471c", fg="white", padx=10, pady=5)
                label_name = Label(frame, text=r_name, bg="#80471c", fg="white", padx=10, pady=5)
                label_cost = Label(frame, text=r_cost, bg="#80471c", fg="white", padx=10, pady=5)
                add_button = Button(frame, text="Add", command=lambda r_name=r_name, 
                            r_cost=r_cost: add_item(f"{r_name} ₹{r_cost}"), bg="#4CAF50", fg="white")
                
                label_id.grid(row=idx, column=0, sticky="w")
                label_name.grid(row=idx, column=1, sticky="w")
                label_cost.grid(row=idx, column=2, sticky="w")
                add_button.grid(row=idx, column=3, sticky="w")
    except Exception as e:
        messagebox.showerror("Error!", f"Error due to {str(e)}", parent=veg)
        return

    # Keep a reference to avoid garbage collection
    veg.bg_image = bg_image

def chinese_food():
    chinese=Toplevel(root)
    chinese.title("chinese Main Course")
    chinese.geometry("700x600")
    chinese.configure(bg="#F08C96")

    # Load the background image
    bg_image = PhotoImage(file="food3.png")

    # Create a label to hold the background image
    bg_label = Label(chinese, image=bg_image)
    bg_label.place(relwidth=1, relheight=1)

    label = Label(chinese, text="Chinese Main Course", font="italic 20 bold", bg="#80471c", fg="white",padx=10, pady=5, borderwidth=1, relief="solid")
    label.pack(pady=10)

    frame = Frame(chinese, bg="#80471c")
    frame.pack(pady=20)

    # Column headers
    header_id = Label(frame, text="I_No", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_name = Label(frame, text="I_Food", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_cost = Label(frame, text="Cost ₹", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    
    header_id.grid(row=0, column=0)
    header_name.grid(row=0, column=1)
    header_cost.grid(row=0, column=2)

    try:
        conn = mysql.connector.connect(host='localhost', user='Your Username', password='Your Password', database='Database name')
        cursor = conn.cursor()
        cursor.execute("select * from chinese;")
        rows = cursor.fetchall()
        conn.close()

        if not rows:
            messagebox.showinfo("Database Empty", "There is no data to show", parent=chinese)
        else:
            for idx, (r_id, r_name, r_cost) in enumerate(rows, start=1):
                label_id = Label(frame, text=r_id, bg="#80471c", fg="white", padx=10, pady=5)
                label_name = Label(frame, text=r_name, bg="#80471c", fg="white", padx=10, pady=5)
                label_cost = Label(frame, text=r_cost, bg="#80471c", fg="white", padx=10, pady=5)
                add_button = Button(frame, text="Add", command=lambda r_name=r_name, 
                            r_cost=r_cost: add_item(f"{r_name} ₹{r_cost}"), bg="#4CAF50", fg="white")
                
                label_id.grid(row=idx, column=0, sticky="w")
                label_name.grid(row=idx, column=1, sticky="w")
                label_cost.grid(row=idx, column=2, sticky="w")
                add_button.grid(row=idx, column=3, sticky="w")
    except Exception as e:
        messagebox.showerror("Error!", f"Error due to {str(e)}", parent=chinese)
        return

    # Keep a reference to avoid garbage collection
    chinese.bg_image = bg_image

def soft_drinks():
    drinks=Toplevel(root)
    drinks.title("Soft drinks")
    drinks.geometry("700x600")
    drinks.configure(bg="#7D7098")

    # Load the background image
    bg_image = PhotoImage(file="food3.png")

    # Create a label to hold the background image
    bg_label = Label(drinks, image=bg_image)
    bg_label.place(relwidth=1, relheight=1)

    label = Label(drinks, text="Soft Drinks", font="italic 20 bold", bg="#80471c", fg="white",padx=10, pady=5, borderwidth=1, relief="solid")
    label.pack(pady=10)

    frame = Frame(drinks, bg="#80471c")
    frame.pack(pady=20)

    # Column headers
    header_id = Label(frame, text="D_No", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_name = Label(frame, text="D_Food", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_cost = Label(frame, text="Cost ₹", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    
    header_id.grid(row=0, column=0)
    header_name.grid(row=0, column=1)
    header_cost.grid(row=0, column=2)

    try:
        conn = mysql.connector.connect(host='localhost', user='Your Username', password='Your Password', database='Database name')
        cursor = conn.cursor()
        cursor.execute("select * from soft_drinks;")
        rows = cursor.fetchall()
        conn.close()

        if not rows:
            messagebox.showinfo("Database Empty", "There is no data to show", parent=drinks)
        else:
            for idx, (r_id, r_name, r_cost) in enumerate(rows, start=1):
                label_id = Label(frame, text=r_id, bg="#80471c", fg="white", padx=10, pady=5)
                label_name = Label(frame, text=r_name, bg="#80471c", fg="white", padx=10, pady=5)
                label_cost = Label(frame, text=r_cost, bg="#80471c", fg="white", padx=10, pady=5)
                add_button = Button(frame, text="Add", command=lambda r_name=r_name, 
                            r_cost=r_cost: add_item(f"{r_name} ₹{r_cost}"), bg="#4CAF50", fg="white")
                
                label_id.grid(row=idx, column=0, sticky="w")
                label_name.grid(row=idx, column=1, sticky="w")
                label_cost.grid(row=idx, column=2, sticky="w")
                add_button.grid(row=idx, column=3, sticky="w")
    except Exception as e:
        messagebox.showerror("Error!", f"Error due to {str(e)}", parent=drinks)
        return

    # Keep a reference to avoid garbage collection
    drinks.bg_image = bg_image

def desserts():
    dessert=Toplevel(root)
    dessert.title("Desserts")
    dessert.geometry("700x600")
    dessert.configure(bg="#987D70")


    # Load the background image
    bg_image = PhotoImage(file="food3.png")

    # Create a label to hold the background image
    bg_label = Label(dessert, image=bg_image)
    bg_label.place(relwidth=1, relheight=1)

    label = Label(dessert, text="Desserts", font="italic 20 bold", bg="#80471c", fg="white",padx=10, pady=5, borderwidth=1, relief="solid")
    label.pack(pady=10)

    frame = Frame(dessert, bg="#80471c")
    frame.pack(pady=20)

    # Column headers
    header_id = Label(frame, text="S_No", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_name = Label(frame, text="S_Food", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_cost = Label(frame, text="Cost ₹", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    
    header_id.grid(row=0, column=0)
    header_name.grid(row=0, column=1)
    header_cost.grid(row=0, column=2)

    try:
        conn = mysql.connector.connect(host='localhost', user='Your Username', password='Your Password', database='Database name')
        cursor = conn.cursor()
        cursor.execute("select * from desserts;")
        rows = cursor.fetchall()
        conn.close()

        if not rows:
            messagebox.showinfo("Database Empty", "There is no data to show", parent=dessert)
        else:
            for idx, (r_id, r_name, r_cost) in enumerate(rows, start=1):
                label_id = Label(frame, text=r_id, bg="#80471c", fg="white", padx=10, pady=5)
                label_name = Label(frame, text=r_name, bg="#80471c", fg="white", padx=10, pady=5)
                label_cost = Label(frame, text=r_cost, bg="#80471c", fg="white", padx=10, pady=5)
                add_button = Button(frame, text="Add", command=lambda r_name=r_name, 
                            r_cost=r_cost: add_item(f"{r_name} ₹{r_cost}"), bg="#4CAF50", fg="white")
                
                label_id.grid(row=idx, column=0, sticky="w")
                label_name.grid(row=idx, column=1, sticky="w")
                label_cost.grid(row=idx, column=2, sticky="w")
                add_button.grid(row=idx, column=3, sticky="w")
    except Exception as e:
        messagebox.showerror("Error!", f"Error due to {str(e)}", parent=dessert)
        return

    # Keep a reference to avoid garbage collection
    dessert.bg_image = bg_image

def salads():
    salad=Toplevel(root)
    salad.title("Salads")
    salad.geometry("700x600")
    salad.configure(bg="#7D7098")

    # Load the background image
    bg_image = PhotoImage(file="food3.png")

    # Create a label to hold the background image
    bg_label = Label(salad, image=bg_image)
    bg_label.place(relwidth=1, relheight=1)

    label = Label(salad, text="Salads", font="italic 20 bold", bg="#80471c", fg="white",padx=10, pady=5, borderwidth=1, relief="solid")
    label.pack(pady=10)

    frame = Frame(salad, bg="#80471c",padx=10, pady=5, borderwidth=1, relief="solid")
    frame.pack(pady=20)

    # Column headers
    header_id = Label(frame, text="Sa_No", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_name = Label(frame, text="Sa_Food", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    header_cost = Label(frame, text="Cost ₹", font="Helvetica 12 bold", bg="#80471c", fg="white", padx=10, pady=5)
    
    header_id.grid(row=0, column=0)
    header_name.grid(row=0, column=1)
    header_cost.grid(row=0, column=2)

    try:
        conn = mysql.connector.connect(host='localhost',user='Your Username', password='Your Password', database='Database name')
        cursor = conn.cursor()
        cursor.execute("select * from salads;")
        rows = cursor.fetchall()
        conn.close()

        if not rows:
            messagebox.showinfo("Database Empty", "There is no data to show", parent=salad)
        else:
            for idx, (r_id, r_name, r_cost) in enumerate(rows, start=1):
                label_id = Label(frame, text=r_id, bg="#80471c", fg="white", padx=10, pady=5)
                label_name = Label(frame, text=r_name, bg="#80471c", fg="white", padx=10, pady=5)
                label_cost = Label(frame, text=r_cost, bg="#80471c", fg="white", padx=10, pady=5)
                add_button = Button(frame, text="Add", command=lambda r_name=r_name, 
                            r_cost=r_cost: add_item(f"{r_name} ₹{r_cost}"), bg="#4CAF50", fg="white")
                
                label_id.grid(row=idx, column=0, sticky="w")
                label_name.grid(row=idx, column=1, sticky="w")
                label_cost.grid(row=idx, column=2, sticky="w")
                add_button.grid(row=idx, column=3, sticky="w")
    except Exception as e:
        messagebox.showerror("Error!", f"Error due to {str(e)}", parent=salad)
        return

    # Keep a reference to avoid garbage collection
    salad.bg_image = bg_image

menu_items = [
    ("Recommended Food", recommended_items),
    ("Indian Veg", indian_veg),
    ("Chinese Food", chinese_food),
    ("Soft Drinks",soft_drinks),
    ("Desserts", desserts),
    ("Salads",salads)
]

menu_frame = tk.Frame(root,bg="#6F4E37")  
menu_frame.pack(side=tk.LEFT,padx=10, pady=10)

# Create and pack buttons in the menu frame
for idx, (name, func) in enumerate(menu_items):
    button = tk.Button(menu_frame, text=name, command=func, width=20, height=2, padx=5, pady=5, borderwidth=1, relief="solid")
    button.pack(pady=5)
        
def get_customer_info():
    # Prompt for customer's name
    customer_name = simpledialog.askstring("Customer Name", "Please enter your name:")
    if not customer_name:
        messagebox.showerror("Input Error", "Customer name is required.")
        return None, None
    
    # Prompt for customer's 10-digit mobile number
    customer_mobile = simpledialog.askstring("Mobile Number", "Please enter your 10-digit mobile number:")
    if not customer_mobile or not re.match(r'^\d{10}$', customer_mobile):
        messagebox.showerror("Input Error", "A valid 10-digit mobile number is required.")
        return None, None
    
    return customer_name, customer_mobile

def print_bill(customer_name, customer_mobile, selected_items, total_cost,gst,grand_total):
    # Generate the bill content
    bill_content = f"Customer Name: {customer_name}\nMobile No: {customer_mobile}\n\n"
    bill_content += f"--------------------------------------------------\n\n"
    bill_content += "Your Order:\n"
    for item in selected_items:
        bill_content += f"{item}\n"
    bill_content += f"---------------------------------------------------\n\n"
    bill_content += f"\n total Bill: ₹{total_cost:.2f}\n"
    bill_content += f"---------------------------------------------------\n\n"
    bill_content += f"\nGST (18%): ₹{gst:.2f}\n"
    bill_content += f"\nGrand Total: ₹{grand_total:.2f}\n"
    bill_content += "\nThank You for visiting our c@nteeN."
    
    # Save the bill content to a text file
    bill_file = "bill.txt"
    with open(bill_file, "w",encoding="utf-8") as file:
        file.write(bill_content)
    
    # Open the text file with Notepad
    os.system(f"notepad.exe {bill_file}")

def order_food():
    # Get customer information
    customer_name, customer_mobile = get_customer_info()
    if customer_name is None or customer_mobile is None:
        return
    
    order = Toplevel(root)
    order.title("Order Food")
    order.geometry("900x750")
    order.resizable(False,False)
    
    # Load the background image
    bg_image = PhotoImage(file="food3.png")

    # Create a label to hold the background image
    bg_label = Label(order, image=bg_image)
    bg_label.place(x=0,y=0,relwidth=1, relheight=1)
    
    
    frame = Frame(order, bg="white")
    frame.place(relx=0.5, rely=0.5, anchor="center")

    # Display customer details in the same frame as the selected items
    customer_info_label = Label(frame, text=f"Customer Name: {customer_name}\nMobile No: {customer_mobile}", font="italic 16 bold", bg="white", fg="black")
    customer_info_label.pack(pady=10)

    label1=Label(frame, text="----------------------------------------------",bg="white", fg="black")
    label1.pack(pady=10)

    label = Label(frame, text="Your Order", font="italic 20 bold", bg="white", fg="black")
    label.pack(pady=10)


    total_cost = 0

    if not selected_items:
        no_order_label = Label(frame, text="No items selected yet",font="italic 15 bold", bg="white", fg="black", padx=10, pady=5)
        no_order_label.pack()
    else:
        for idx, item in enumerate(selected_items, start=1):
            item_label = Label(frame, text=item, bg="white",font="italic 15 bold", fg="black", padx=10, pady=5)
            item_label.pack()

            # Extract the cost from the item string and add to total_cost
            cost = float(re.search(r'₹(\d+(?:\.\d+)?)', item).group(1))
            total_cost += cost
        label1=Label(frame, text="----------------------------------------------",bg="white", fg="black")
        label1.pack(pady=10)
        
        # Display total cost in the same frame as the selected items
        total_cost_label = Label(frame, text=f"total Bill: ₹{total_cost:.2f}", font="italic 20 bold", bg="white", fg="black")
        total_cost_label.pack(pady=10)
        
        # Calculate GST
        gst = total_cost * 0.18
        grand_total = total_cost + gst
    
        # Display GST amount
        gst_label = Label(frame, text=f"GST (18%): ₹{gst:.2f}", font="italic 15 bold", bg="white", fg="black")
        gst_label.pack(pady=5)
    
        # Display grand total
        grand_total_label = Label(frame, text=f"Grand Total: ₹{grand_total:.2f}", font="italic 20 bold", bg="white", fg="black")
        grand_total_label.pack(pady=10)

        label2=Label(frame, text="----------------------------------------------",bg="white", fg="black")
        label2.pack(pady=10)
    
        info_label=Label(frame, text="Thank You for visiting our c@nteeN.", font="italic 16 bold", bg="white", fg="black")
        info_label.pack(pady=10)

        print_bill_button = Button(frame, text="Print Bill", font="italic 15 bold", bg="white", fg="black",
                                   command=lambda: print_bill(customer_name, customer_mobile, selected_items, total_cost,gst,grand_total))
        print_bill_button.pack(pady=20)
    exit_btn=Button(order, text="Exit", command=order.destroy,
                   font="italic 15 bold", bg="white", fg="black")
    exit_btn.place(x=800,y=700)
    
    # Keep a reference to avoid garbage collection
    order.bg_image = bg_image

def message_canteen():
    message = Toplevel(root)
    message.title("Message for Our C@nteeN")
    message.geometry("600x500")
    # Load the background image
    bg_image = PhotoImage(file="food3.png")

    # Create a label to hold the background image
    bg_label = Label(message, image=bg_image)
    bg_label.place(relwidth=1, relheight=1)
    
    label = Label(message, text="Message for Our C@nteeN", font="italic 20 bold", bg="white", fg="black",padx=10, pady=5, borderwidth=1, relief="solid")
    label.pack(pady=10)

    def submit_review():
        name=entry_name.get()
        message=entry_message.get()
        if (name=="" and message==""):
            messagebox.showerror("Error","Please Fill the entries")
        else:
            try:
                #connect to mysql database
                conn= mysql.connector.connect(host='localhost',user='Your Username', password='Your Password', database='Database name')
                if conn.is_connected():
                    cursor= conn.cursor()
        
                    #insert data
                    sql= "insert into review (name,message) values(%s,%s)"
                    val=(name,message)
                    cursor.execute(sql,val)
                    conn.commit()
                    messagebox.showinfo("Success","Data inserted successfully!")
                    reset_fields()
                    exit()
                else:
                    messagebox.showerror("Error","Failed to connect to the database.")
            except Error as e:
                messagebox.showerror("Error",f"Error:{e}")
            finally:
                if conn.is_connected():
                    cursor.close()
                    conn.close()
            
    label_name=tk.Label(message,text="Name:",font="italic ",padx=5, pady=5, borderwidth=1, relief="solid")
    label_name.place(x=80,y=70)
    
    entry_name=tk.Entry(message,font="italic ")
    entry_name.place(x=200,y=70)
    
    label_message=tk.Label(message,text="Message:",font="italic ",padx=5, pady=5, borderwidth=1, relief="solid")
    label_message.place(x=80,y=120)
    
    entry_message=tk.Entry(message,font="italic ")
    entry_message.place(x=200,y=120)

    def reset_fields():
        entry_name.delete(0, END)
        entry_message.delete(0, END)

    # Keep a reference to avoid garbage collection
    message.bg_image = bg_image

    btn= Button(message, text= "Submit",font="italic ", command= submit_review,padx=10, pady=5, borderwidth=1, relief="solid").place(x=80,y=170)

def review_canteen():
    review = Toplevel(root)
    review.title("C@nteeN ReVieW")
    review.geometry("600x500")

    # Load the background image
    bg_image = PhotoImage(file="food3.png")

    # Create a label to hold the background image
    bg_label = Label(review, image=bg_image)
    bg_label.place(relwidth=1, relheight=1)
    
    label = Label(review, text="C@nteeN ReView frOm CustOmer$", font="italic 20 bold", bg="white", fg="black",padx=10, pady=5, borderwidth=1, relief="solid")
    label.pack(pady=10)
    
    try:
        conn= mysql.connector.connect(host='localhost',user='Your Username', password='Your Password', database='Database name')
        cursor=conn.cursor()
        conn.commit()
        cursor.execute("select * from review;")
        rows=cursor.fetchall()
        if rows == None:
            messagebox.showinfo("Database Empty","There is no data to show",parent=review)
            conn.close()
        else:
            frame = tk.Frame(review, bg="#C19A6B")
            frame.pack(pady=20)

            for idx, row in enumerate(rows):
                for col_idx, col in enumerate(row):
                    entry_label = Label(frame, text=col, bg="white", fg="black", padx=10, pady=5, borderwidth=1, relief="solid")
                    entry_label.grid(row=idx, column=col_idx, padx=5, pady=5)
            conn.close()
    except Exception as e:
        messagebox.showerror("Error!",f"Error due to {str(e)}",parent=review)
        
    # Keep a reference to avoid garbage collection
    review.bg_image = bg_image
    
btn2=tk.Button(label17,text="Your Order List",command=order_food,width=20, height=2,padx=10, pady=5, borderwidth=1, relief="solid")
btn2.place(x=280,y=330)

btn3=tk.Button(label17,text="Message for Our C@nteeN",command=message_canteen,width=20, height=2,padx=10, pady=5, borderwidth=1, relief="solid")
btn3.place(x=800,y=360)

btn4=tk.Button(label17,text="C@nteeN ReVieW",command=review_canteen,width=20, height=2,padx=10, pady=5, borderwidth=1, relief="solid")
btn4.place(x=800,y=420)

btn5=tk.Button(label17, text="Exit", command=root.destroy,width=10,height=2,padx=10, pady=5, borderwidth=1, relief="solid").place(x=840, y=500)
# Run the application
root.mainloop()
```
